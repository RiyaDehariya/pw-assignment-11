```python
# 1) What is the distribution of age among heart failure patients in the dataset ?
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df.age, kde = True, color = "skyblue")
plt.title("Distrbution of age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()



#Insights
#The mean age of person having heart failure is around 50 to 60
```

```python
2. How does the death rate vary with age?
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data=df, x='age', weights='DEATH_EVENT', bins=30, kde=True, color='blue')
plt.title('Variation of Death rate with age')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Death Rate (Density)', fontsize=14)
plt.grid(True)
plt.show()

#Insights
#From below graph we see that The patients above 60 yrs of age has higher chances of death
#Here 0 indicates survival and 1 indicates death.
```

```python
3. What is the percentage of male and female patients in the dataset?
```

```python
df['sex'].value_counts().plot.pie(autopct='%1.1f%%', explode=(0,0.1), labels=('males','females'), startangle=90)
plt.title('Percentage of Male and Female Patients')


#Insights
#Percentage of females having heart failure is 35.1%
#percentage of males having heart failure 64.9%
#we can interpret that males have high chance of heart attack as compared to females.
```

```python
# 4. How does the platlates count vary among different age groups?
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(data = df, x = 'age', y = 'platelets')
plt.title('Variation of Platelets count with Age')
plt.xlabel('Age')
plt.ylabel('Platelets')
plt.show()

#Insights
#We can see that Count of platelets decreases as age increases
```

```python
5. Is there a corelation between creatinine and sodium levels in blood?
```

```python
df1 = df[['serum_creatinine', 'serum_sodium']]
df1.corr()
```

```python
sns.heatmap(df1.corr(), annot = True, cmap = 'coolwarm')
plt.title('Correlation between creatinine and sodium level')

#Insights
#There is negative correlation between ceatinne and sodium levels.
```

```python
6.How does the prevalence of high blood pressure differ between male and female patients?
```

```python
cross_tab = pd.crosstab(df['sex'], df['high_blood_pressure'])
cross_tab
```

```python
cross_tab.plot(kind = 'bar', stacked = True, color = ['#6DA7DE', '#FF9A8B'], figsize = (10, 6))
plt.xlabel("Gender")
plt.ylabel("count")
plt.legend(title="High blood pressure")
plt.title("Prevalance of High blood pressure with respect to male and female")
plt.show()

#Insights
#Here 0 represents female and 1 represents male in x axis
#From below graph we can interpret that among 105 females, approximately 45 females have high blood pressure
#Among 194 males, approximately 60 males have high blood pressure.
```

```python
7. What is the relationship betweeen smoking habits and occurence of heart failure?
```

```python
#sns.countplot(df.smoking)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='smoking')
plt.title('Relationship between Smoking Habits and Occurrence of Heart Failure', fontsize=16)
plt.xlabel('Smoking Habit (0 = Non-smoker, 1 = Smoker)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Heart Failure (0 = No, 1 = Yes)', loc='upper right')
plt.grid(True)
plt.show()
#Insights
#Here 0 indicates no smoking and 1 indicates smoking
#we can see from graph that poeple who smoke has high chance of heart failure bu people who do not smoke also has risk of heart failure
```

```python
8. Are there any noticeable patterns in the distribution of death events across different age groups
```

```python
plt.figure(figsize = (10, 6))
sns.boxplot(x = 'DEATH_EVENT', y='age', data=df)
plt.xlabel('Death Event')
plt.ylabel('Age')
plt.title("Boxplot of Death event by age")
plt.show()

#person having age between 55 and 75 have high tendency of death
```

```python
9. Is there any significant difference in ejection_fraction between patients with and without diabetes?
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='diabetes', y='ejection_fraction', palette='Set2')
plt.title('Ejection Fraction Distribution for Patients with and without Diabetes', fontsize=16)
plt.xlabel('Diabetes (0 = No, 1 = Yes)', fontsize=14)
plt.ylabel('Ejection Fraction', fontsize=14)
plt.grid(True)
plt.show()
```
